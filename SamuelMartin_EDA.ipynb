# Looking into the Movies_gross CSV file

```python
df = pd.read_csv('data/bom.movie_gross.csv.gz')
df.head()
# df.shape
```

From running the code above, 3387 rows of movies, with columns like title, gross profit, and stuff like that

```python
### Looking for repeats in titles
titles = df.title.unique()
len(titles)
### There are no repeats
```

```python
df.info()
```

```python
### Adding on a total revenue column which is the sum or domestic and foreign
### Change type of foreign_gross to int
df['foriegn_int'] = pd.to_numeric(df.foreign_gross.str.replace(',',''))

df['Total_Gross_Millions'] = (df.domestic_gross + df.foriegn_int) / 1000000
df.head()
```

## Findings
- Would be a good file to use for basic analysis comparing yearly growth of movie revenue
- Very clean
