# Shifting through directors and Writers on the IMDB DB

```python
### Import the nessesary libraries
import pandas as pd
import sqlite3 
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import scipy.stats as stats
import seaborn as sns
```

```python
conn = sqlite3.connect('data/sqldata/im.db')
```

```python
#Average movie ratings
q = """
SELECT AVG(averagerating)
FROM movie_ratings
;
"""
pd.read_sql(q, conn)
```

```python
#Select all writers, directors, and actors who were in movies with ratings above the average
q = """
SELECT
    primary_name AS name,
    primary_title AS p_title,
    original_title AS o_title,
    genres,
    job,
    averagerating AS rating
FROM persons
JOIN principals
    USING(person_id)
JOIN movie_basics
    USING(movie_id)  
JOIN movie_ratings
    USING(movie_id)
WHERE
    averagerating > 6.3
    AND
    job != 'None'
ORDER BY
    rating DESC
;
"""
df = pd.read_sql(q, conn)
```
